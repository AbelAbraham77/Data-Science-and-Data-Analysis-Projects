<h1>OLYMPICS DATA ANALYSIS</h1>

<h3>Importing All Libraries </h3>

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

<h3>Importing and Reading Data</h3>

```python
data = pd.read_csv('athlete_events.csv')

print(data.head(), data.describe(), data.info())
```

<h3>Merging Two Data Frames</h3>

<p>Merging the two dataframes using pandas.merge()</p>

```python
regions = pd.read_csv('datasets_31029_40943_noc_regions.csv')
print(regions.head())

merged = pd.merge(data, regions, on='NOC', how='left')
print(merged.head())
```

<h3>Data Analysis of Gold Medallists</h3>
<p>Creating a new data frame with only gold medallists</p>

```python
goldMedals = merged[(merged.Medal == 'Gold')]
print(goldMedals.head())
```

<h3>Gold Medallists classified on the basis of age</h3>
<p>Here we are creating a graph of the number of gold medals with respect to age. In the graph, the x-axis represents the age of the players and the y-axis represents the number of gold medals.</p>

```python
plt.figure(figsize=(20, 10))
plt.title('Distribution of Gold Medals')
sns.countplot(goldMedals['Age'])
plt.show()
```

```python
goldMedals = merged[(merged.Medal == 'Gold')]
print('The no of athletes is',
      goldMedals['ID'][goldMedals['Age'] > 50].count(), '\n')
print(goldMedals[goldMedals['Age'] > 50])
```

<p>Create a new dataframe in which we insert the new set of people and create a data visualisation with it.</p>

```python
masterDisciplines = goldMedals['Sport'][goldMedals['Age'] > 50]
plt.figure(figsize=(20, 10))
plt.tight_layout()
sns.countplot(masterDisciplines)
plt.title('Gold Medals for Athletes Over 50')
plt.show()
```

<p>Now we display all women athletes who have played in the summer olympic games and show the increase in women athletes after a long period via graphical representation.</p>

```python
womenInOlympics = merged[(merged.Sex == 'F') &
						(merged.Season == 'Summer')]
print(womenInOlympics.head(10))

sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenInOlympics)
plt.title('Women medals per edition of the Games')
plt.show()
```

<h3>Top 5 Countries</h3>
<p>Now we are going to display the top 5 countries using catplot.</p>

```python
print(goldMedals.region.value_counts().reset_index(name='Medal').head())

totalGoldMedals = goldMedals.region.value_counts().reset_index(name='Medal').head(5)
g = sns.catplot(x = "region", y = "Medal", data=totalGoldMedals, height=6, kind="bar", palette="muted")
g.despine(left=True)
g.set_xlabels("Top 5 countries")
g.set_ylabels("Number of Medals")
plt.title('Medals per Country')
plt.show()
```

<h3>Weight Analysis of Athletes</h3>
<p>We are going to see how the weights of male weight-lifters vary over the years using pointplot.</p>

```python
MenOverTime = merged[(merged.Sex == 'M') & (merged.Season == 'Summer')]
wlMenOverTime = MenOverTime.loc[MenOverTime['Sport'] == 'Weightlifting']

plt.figure(figsize=(20, 10))
sns.pointplot(x = 'Year', y = 'Weight', data=wlMenOverTime, palette='Set2')
plt.title('Weight over year for Male Lifters')
plt.show()
```
